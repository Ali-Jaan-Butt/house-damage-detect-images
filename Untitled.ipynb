```python
from matplotlib import pyplot as plt
roles = ['Entry-level', 'Mid-level', 'Senior-level', 'Leadership']
percentages = [25, 15, 10, 5]

# Create the bar chart
plt.figure(figsize=(15, 6))
plt.bar(roles, percentages, color='skyblue')
plt.xlabel('Blockchain Roles')
plt.ylabel('Percentage of Women')
plt.title('Percentage of Women in Various Blockchain Roles')
plt.ylim(0, 30)  # Adjust the y-axis limit for better visualization
plt.savefig("Graph.jpg")
# Display the chart
plt.show()



# 2. HERE IS THE CODE FOR THE SECOND CHART.

# Quotes
quotes = [
    ("On Racial and Gender Biases", "As an African American woman, I often feel like I have to prove myself twice as much as my male counterparts to gain the same level of respect and recognition."),
    ("On Trust and Transparency", "Decisions were made behind closed doors, and I often felt excluded from important discussions. Transparency is key to building trust and ensuring everyone feels valued."),
    ("On Mentorship Programs", "The mentorship program was a game-changer for me. It provided me with the support and guidance I needed to advance my career in blockchain.")
]

# Create the figure
plt.figure(figsize=(15, 8))

# Add quotes to the figure
for i, (title, quote) in enumerate(quotes):
    plt.text(0.1, 0.8 - i * 0.3, f"{title}:\n\n\"{quote}\"", fontsize=12, wrap=True)

# Remove axes
plt.axis('off')
plt.title('Direct Quotes from Participants', fontsize=14)
plt.savefig("Qoutes.jpg")
# Display the figure
plt.show()
```
